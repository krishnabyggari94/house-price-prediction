#### Feature Selection

```python
import pandas as pd
import numpy as np
```

```python
x_train=pd.read_csv('x_train.csv')
x_test=pd.read_csv('x_test.csv')
y_train=pd.read_csv('y_train.csv')
y_test=pd.read_csv('y_test.csv')
x_train.head()
```

```python
from sklearn.tree import ExtraTreeRegressor
tree_reg=ExtraTreeRegressor()
tree_reg.fit(x_train,y_train)
```

```python
tree_reg.feature_importances_
```

```python
sum(tree_reg.feature_importances_)
```

```python
print(len(tree_reg.feature_importances_))
print([val for val in tree_reg.feature_importances_ if val>0.005])
print(len([val for val in tree_reg.feature_importances_ if val>0.005]))
print(sum([val for val in tree_reg.feature_importances_ if val>0.005]))
```

```python
selected_cols=x_train.columns[tree_reg.feature_importances_>0.005]
selected_cols
```

```python
x_train=x_train[selected_cols]
x_test=x_test[selected_cols]
x_train.head()
```

#### Model Building

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
lin_model=LinearRegression()
lin_model.fit(x_train,y_train)
y_pred=lin_model.predict(x_test)
lin_model_r2_score=r2_score(y_test,y_pred)
print('lin_model_r2_score is :',lin_model_r2_score) 
```

```python
from sklearn.metrics import mean_squared_error,mean_squared_log_error
from math import sqrt
lin_model_rmle_score=sqrt(mean_squared_log_error(y_test,y_pred))
print('lin_model_rmle_score is :',lin_model_rmle_score) 
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV,KFold
folds=KFold(n_splits=5)
model=Ridge(random_state=42)
params={'alpha':[0.0001,0.001,0.005,0.01,0.05,0.1,1,2,4,5],'solver':['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga']}
ridge_cv=GridSearchCV(model,param_grid=params,cv=folds)
ridge_cv.fit(x_train,y_train)
y_pred=ridge_cv.predict(x_test)
ridge_model_r2_score=r2_score(y_test,y_pred)
print('ridge_model_r2_score is :',ridge_model_r2_score) 
ridge_model_rmle_score=sqrt(mean_squared_log_error(y_test,y_pred))
print('ridge_model_rmle_score is :',ridge_model_rmle_score)
print(ridge_cv.best_estimator_)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV,KFold
folds=KFold(n_splits=5)
model=Lasso(random_state=42)
params={'alpha':[0.0001,0.001,0.005,0.01,0.05,0.1,1,2,4,5]}
lasso_cv=GridSearchCV(model,param_grid=params,cv=folds)
lasso_cv.fit(x_train,y_train)
y_pred=lasso_cv.predict(x_test)
lasso_model_r2_score=r2_score(y_test,y_pred)
print('lasso_model_r2_score is :',lasso_model_r2_score) 
lasso_model_rmle_score=sqrt(mean_squared_log_error(y_test,y_pred))
print('lasso_model_rmle_score is :',lasso_model_rmle_score)
print(lasso_cv.best_estimator_)
```

```python
from sklearn.neighbors import KNeighborsRegressor
folds=KFold(n_splits=5)
model=KNeighborsRegressor(n_jobs=-1)
params={'n_neighbors':[3,5,6,7,9,10],'weights':['uniform','distance'],'p':[0.5,1,1.5,2]}
knn_cv=GridSearchCV(model,param_grid=params,cv=folds)
knn_cv.fit(x_train,y_train)
y_pred=knn_cv.predict(x_test)
knn_model_r2_score=r2_score(y_test,y_pred)
print('knn_model_r2_score is :',knn_model_r2_score) 
knn_model_rmle_score=sqrt(mean_squared_log_error(y_test,y_pred))
print('knn_model_rmle_score is :',knn_model_rmle_score)
print(knn_cv.best_estimator_)
```

```python
from sklearn.svm import SVR
folds=KFold(n_splits=5)
model=SVR()
params={'kernel':['poly','rbf'],'degree':[2,3],'C':np.arange(0,1100,100)}
svr_cv=GridSearchCV(model,param_grid=params,cv=folds)
svr_cv.fit(x_train,y_train)
y_pred=svr_cv.predict(x_test)
svr_model_r2_score=r2_score(y_test,y_pred)
print('svr_model_r2_score is :',svr_model_r2_score) 
svr_model_rmle_score=sqrt(mean_squared_log_error(y_test,y_pred))
print('svr_model_rmle_score is :',svr_model_rmle_score)
print(svr_cv.best_estimator_)
```

```python
from sklearn.tree import DecisionTreeRegressor
folds=KFold(n_splits=5)
model=DecisionTreeRegressor()
params={'criterion':['mse','friedman_mse'],'max_depth':[5,6,7,8,9,10],'min_samples_split':np.arange(5,20,3),'min_samples_leaf':np.arange(5,20,3)}
tree_cv=GridSearchCV(model,param_grid=params,cv=folds)
tree_cv.fit(x_train,y_train)
y_pred=tree_cv.predict(x_test)
tree_model_r2_score=r2_score(y_test,y_pred)
print('tree_model_r2_score is :',tree_model_r2_score) 
tree_model_rmle_score=sqrt(mean_squared_log_error(y_test,y_pred))
print('tree_model_rmle_score is :',tree_model_rmle_score)
print(tree_cv.best_estimator_)
```

```python
from sklearn.ensemble import RandomForestRegressor
folds=KFold(n_splits=5)
model=RandomForestRegressor(random_state=42)
params={'max_features':['auto','sqrt','log2'],'min_samples_split':np.arange(5,20,3),'min_samples_leaf':np.arange(5,20,3)}
rf_cv=GridSearchCV(model,param_grid=params,cv=folds)
rf_cv.fit(x_train,y_train)
y_pred=rf_cv.predict(x_test)
rf_model_r2_score=r2_score(y_test,y_pred)
print('rf_model_r2_score is :',rf_model_r2_score) 
rf_model_rmle_score=sqrt(mean_squared_log_error(y_test,y_pred))
print('rf_model_rmle_score is :',rf_model_rmle_score)
print(rf_cv.best_estimator_)
```
