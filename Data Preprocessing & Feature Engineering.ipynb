```python
all_cols=df.columns
print([col for col in all_cols if df[col].isnull().sum()>0 and df[col].dtypes=='O'])
print([col for col in all_cols if df[col].isnull().sum()>0 and df[col].dtypes!='O'])
```

```python
missing_num_cols=[col for col in all_cols if df[col].isnull().sum()>0 and df[col].dtypes!='O']
missing_cat_cols=[col for col in all_cols if df[col].isnull().sum()>0 and df[col].dtypes=='O']
print(len(missing_cat_cols))
print(len(missing_num_cols))
```

```python
for col in missing_num_cols:
    print('median of {} column is {}'.format(col,df[col].median()))
```

```python
for col in missing_num_cols:
    df[col+'_missing']=np.where(df[col].isnull(),1,0)
    df[col].fillna(df[col].median(),inplace=True)
df.head()
```

```python
for col in missing_cat_cols:
    df[col].fillna('Missing',inplace=True)
df.head()
```

```python
df.isnull().sum().sum()
```

```python
year_cols=['YearBuilt','YearRemodAdd','GarageYrBlt','YrSold']
for col in year_cols:
    if col!='YrSold':
        df[col]=df['YrSold']-df[col]
df.head()
```

```python
cat_cols=[col for col in df.columns if df[col].dtypes=='O']
num_cols=[col for col in df.columns if df[col].dtypes!='O']
print(len(cat_cols))
print(len(num_cols))
```

```python
for col in cat_cols:
    chk=[]
    samp_dict=dict(df[col].value_counts())
    #print(samp_dict.keys())
    #print(samp_dict.values())
    for val in samp_dict.keys():
        if samp_dict[val]<14:
            df[col]=np.where(df[col]==val,'Rare_feat',df[col])
df.head()
```

```python
for col in num_cols:
    upper=np.quantile(df[col],0.995)
    lower=np.quantile(df[col],0.005)
    df[col]=np.where(df[col]>upper,upper,df[col])
    df[col]=np.where(df[col]<lower,lower,df[col])
df.head()
```

```python
for val in num_cols:
    if val!='Id' and val!='SalePrice':
        if 0 in df[val].unique():
            pass
        else:
            df[val]=np.log(df[val])
df.head()
```

```python
for col in cat_cols:
    chk=dict(df.groupby(col)['SalePrice'].mean().sort_values())
    df[col]=df[col].map(chk)
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop(labels=['Id','YrSold','SalePrice'],axis=1),df[['SalePrice']],test_size=0.2,random_state=42)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
transformed_x_train=scaler.transform(x_train)
transformed_x_test=scaler.transform(x_test)
```
