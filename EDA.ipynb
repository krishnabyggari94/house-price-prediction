### House Price Prediction EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('train.csv')
pd.set_option("display.max_columns", None)
df.head()
```

```python
print(df.shape)
```

```python
all_columns=df.columns
print(len(all_columns))
```

```python
df.describe()
```

```python
df.info()
```

```python
x=df.drop(labels=['Id'],axis=1)
x.head()
```

```python
x.isnull().sum().sum()
```

```python
num_cols=[col for col in x.columns if x[col].dtypes!='O']
print(len(num_cols))
```

```python
print(num_cols)
```

```python
year_cols=['YearBuilt','GarageYrBlt','YearRemodAdd','YrSold']
num_cols=list(set(num_cols)-set(year_cols))
print(num_cols)
```

```python
cat_cols=[col for col in x.columns if x[col].dtypes=='O']
print(len(cat_cols))
print(cat_cols)
```

```python
missing_cols=[col for col in x.columns if x[col].isnull().sum()!=0]
print(len(missing_cols))
print(missing_cols)
```

```python
x[missing_cols].isnull().sum()
```

```python
for col in missing_cols:
    data=x.copy()
    data[col]=np.where(data[col].isnull(),1,0)
    data.groupby(col)['SalePrice'].median().plot(kind='bar')
    plt.show()
```

```python
x[year_cols].head()
```

```python
for col in year_cols:
    data=x.copy()
    data.groupby(col)['SalePrice'].median().plot()
    plt.show()
```

```python
for col in year_cols:
    if col!='YrSold':
        data=x.copy()
        data['YrSold-'+col]=data['YrSold']-data[col]
        data.groupby('YrSold-'+col)['SalePrice'].median().plot()
        plt.show()
```

```python
x[num_cols].head()
```

```python
discrete_cols=[col for col in num_cols if x[col].nunique()<20]
print(len(discrete_cols))
print(discrete_cols)
```

```python
for col in discrete_cols:
    data=x.copy()
    data.groupby(col)['SalePrice'].median().plot(kind='bar')
    #sns.barplot()
    plt.show()
```

```python
cont_cols=[col for col in num_cols if col not in discrete_cols]
print(len(cont_cols))
print(cont_cols)
```

```python
for col in cont_cols:
    data=x.copy()
    data[col].hist(bins=30)
    plt.xlabel(col)
    plt.show()
```
